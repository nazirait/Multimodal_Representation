```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
```

```python
path = "src/comparative/checkpoints/lightning_logs/movielens_vilbert/version_0/events.out.tfevents.1762699226.ws-zelazny.714801.0"
```

```python
ea = EventAccumulator(path) # event file
ea.Reload()

print("Scalars:", ea.Tags().get('scalars'))
print("Histograms:", ea.Tags().get('histograms'))
print("Images:", ea.Tags().get('images'))
print("Tensors:", ea.Tags().get('tensors'))

# ex: some scalar values
for tag in ea.Tags().get('scalars', []):
    print(f"\n=== {tag} ===")
    events = ea.Scalars(tag)
    for ev in events[:5]:   # first 5 entries
        print(ev.step, ev.value)
```
