```python
import matplotlib.pyplot as plt

top_10_labels = [
    ("man", 95375),
    ("person", 70765),
    ("window", 69335),
    ("tree", 49549),
    ("building", 47340),
    ("shirt", 46382),
    ("wall", 45039),
    ("woman", 42081),
    ("sign", 40714),
    ("sky", 40049),
]

labels, counts = zip(*top_10_labels[:10])

print(labels)
print(counts)
```

```python
import seaborn as sns


sns.set_theme(style="whitegrid", font_scale=1.1)

plt.figure(figsize=(8, 6))

ax = sns.barplot(
    x=counts,
    y=labels,
    palette="viridis",   
    edgecolor="none"
)

ax.invert_yaxis()

ax.set_xlabel("Frequency")
ax.set_ylabel("Object Category")
ax.set_title("Top-10 Object Categories")

sns.despine(left=True, bottom=False)
plt.tight_layout()
plt.show()
```
